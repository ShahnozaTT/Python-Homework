```python
#     - Percentage of population for each salary category;

import pandas as pd
import numpy as np

# Границы интервалов (bins)
bins = [0, 200000, 400000, 600000, 800000, 1000000,
        1200000, 1400000, 1600000, 1800000, np.inf]

# Названия категорий (labels)
labels = [
    "till $200,000",
    "$200,001 - $400,000",
    "$400,001 - $600,000",
    "$600,001 - $800,000",
    "$800,001 - $1,000,000",
    "$1,000,001 - $1,200,000",
    "$1,200,001 - $1,400,000",
    "$1,400,001 - $1,600,000",
    "$1,600,001 - $1,800,000",
    "$1,800,001 and over"
]

import sqlite3

# Загружаем данные из базы
conn = sqlite3.connect("population.db")
df = pd.read_sql('SELECT id, salary, state FROM population', conn)





# Присваиваем Salary Band
df["Salary Band"] = pd.cut(df["salary"], bins=bins, labels=labels, right=True)

print(df.head())


```

```python
#     - Average salary in each salary category;

avg_salary = df.groupby("Salary Band")["salary"].mean()
print(avg_salary)
```

```python
#     - Median salary in each salary category;

median_salary = df.groupby("Salary Band")["salary"].median()
print(median_salary)
```

```python
#     - Number of population in each salary category;

num_population = df.groupby("Salary Band")['id'].count()


total_population = len(df)

# Доля в процентах
percent_population = (num_population / total_population) * 100

print(percent_population)
''
```

```python
result = pd.DataFrame({
    "Population %": percent_population,
    "Average Salary": avg_salary,
    "Median Salary": median_salary,
    "Count": num_population
})

print(result)
```

```python
# 3. Calculate the same measures in each State (Note: Use SQL only to select data from database. All the other calculations should be done in python.)

avg_salary_by_state = df.groupby(['state', 'Salary Band'])['salary'].mean()


median_salary_by_state = df.groupby(['state', 'Salary Band'])['salary'].median()


num_population_by_state = df.groupby(['state', 'Salary Band'])['id'].count() 
num_population_by_state

total_population_by_state = len(df)

# # Доля в процентах
percent_population_by_state = (num_population_by_state / total_population_by_state) * 100

print(percent_population_by_state)
# ''
```

```python
result_by_state= pd.DataFrame({
    "Population % by state": percent_population_by_state,
    "Average Salary by state": avg_salary_by_state,
    "Median Salary by state": median_salary_by_state,
    "Count by state": num_population_by_state
})

print(result_by_state)
```
